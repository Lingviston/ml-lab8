```python
y, x = svm_read_problem(learn_scaled_converted_data_set_file_path)

k = 20

d_values = range(1, 5)
C_values = [2**x for x in range(-k, k + 1)]

learn_problem  = svm_problem(y, x)
for d in d_values:
    for C in C_values:
        params_string = '-v 10 -t 1 -d {0} -c {1}'.format(d, C)
        
        print(params_string)
        learn_parameter = svm_parameter(params_string)
        m = svm_train(learn_problem, learn_parameter)
```
